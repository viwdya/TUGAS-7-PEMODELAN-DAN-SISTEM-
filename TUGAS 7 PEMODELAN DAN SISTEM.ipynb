```python
import numpy as np
import matplotlib.pyplot as plt

# Definisi kategori
states = ["Rendah", "Sedang", "Tinggi", "Sangat Tinggi"]
n_states = len(states)

transition_matrix = np.array([
    [0.5, 0.3, 0.15, 0.05],   # Probabilitas penghasilan Rendah
    [0.2, 0.4, 0.3, 0.1],     # Probabilitas penghasilan Sedang
    [0.1, 0.3, 0.4, 0.2],     # Probabilitas penghasilan Tinggi
    [0.05, 0.15, 0.3, 0.5]    # Probabilitas penghasilan Sangat Tinggi
])

# Jumlah hari untuk simulasi
n_days = 30
initial_state = 1

income_history = [initial_state]
for _ in range(n_days - 1):
    current_state = income_history[-1]
    next_state = np.random.choice(n_states, p=transition_matrix[current_state])
    income_history.append(next_state)
days = range(1, n_days + 1)
income_labels = [states[state] for state in income_history]

# hasil
plt.figure(figsize=(12, 6))
plt.plot(days, income_history, marker='o', linestyle='-', color='b')
plt.yticks(range(n_states), states)
plt.xlabel("Hari")
plt.ylabel("Kategori Penghasilan")
plt.title("Simulasi Penghasilan Harian Toko Online Menggunakan Markov Chain")
plt.grid(True)
plt.show()



```
